```python
import pandas as pd
```

```python
movies_rating = pd.read_csv('ratings.csv')
movies_rating.head(4)
```

```python
movies_names = pd.read_csv('movies.csv')
movies_names.head(4)
```

```python
full_info = movies_rating.merge(movies_names, on='movieId', how='left')
full_info.head(5)
```

```python
five_stars_movies = full_info[full_info['rating'] == 5]
five_stars_movies
```

```python
five_stars_movies['title'].value_counts().head(10)
```

```python
power = pd.read_csv('power.csv')
power.head(5)
```

```python
pribalt_power = power[((power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] == 'Estonia'))
&((power['year'] > 2004)&(power['year'] < 2011)&(power['quantity'] >= 0))]
pribalt_power.head(10)
```

```python
cat_pribalt_power = pribalt_power[(pribalt_power['category'] == 4)|(pribalt_power['category'] == 12)|(pribalt_power['category'] == 21)]
cat_pribalt_power['quantity'].sum()
```
